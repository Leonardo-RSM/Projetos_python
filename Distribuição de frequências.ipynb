# Distribuição de Frequências

```python
import numpy as np
import pandas as pd
from collections import Counter
enem_sp = pd.read_csv('/Users/paull/OneDrive/Área de Trabalho/LEONARDO/python/jupyter/enem_2019_tratado.csv',
                      sep = ',', encoding = 'iso-8859-1')
```

```python
enem_sp.head()
```

```python
enem_sorocaba = enem_sp.loc[enem_sp.NO_MUNICIPIO_RESIDENCIA == 'Sorocaba']
enem_sorocaba.head()
```

```python
enem_sorocaba.shape
```

```python
idades = enem_sorocaba['IDADE']
```

```python
freq_absoluta = Counter(idades)
freq_absoluta
```

```python
freq_absoluta = pd.DataFrame.from_dict(freq_absoluta)
freq_absoluta
```

```python
freq_absoluta = freq_absoluta.sort_index(ascending=True)
freq_absoluta
```

```python
freq_relativa = freq_absoluta/ freq_absoluta.sum()
```

```python
freq_relativa
```

Frequencia relativa é a divisão entre a freqeuncia absoluta de um dado e a soma de todos os dados da amostra

```python
freq_relativa_percentual = freq_relativa*100
freq_relativa_percentual
```
