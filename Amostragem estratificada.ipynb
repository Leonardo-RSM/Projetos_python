dividir a amostra em grupos e depois selcionamos aleatoriamente uma amostra de cada grupo

```python
import pandas as pd
import numpy as np
```

```python
enem_sp = pd.read_csv('/Users/paull/OneDrive/Área de Trabalho/LEONARDO/python/jupyter/enem_2019_tratado.csv',
                     sep = ',', encoding = 'iso-8859-1')
enem_campinas = enem_sp.loc[enem_sp.NO_MUNICIPIO_RESIDENCIA == 'Campinas']
```

```python
enem_campinas['RACA'].value_counts()
```

```python
branca = enem_campinas.loc[enem_campinas.RACA == 'branca'].sample(frac=0.10)
branca.shape
```

```python
parda = enem_campinas.loc[enem_campinas.RACA == 'parda'].sample(frac=0.10)
preta = enem_campinas.loc[enem_campinas.RACA == 'preta'].sample(frac=0.10)
amarela = enem_campinas.loc[enem_campinas.RACA == 'amarela'].sample(frac=0.10)
indigena = enem_campinas.loc[enem_campinas.RACA == 'indígena'].sample(frac=0.10)
nd = enem_campinas.loc[enem_campinas.RACA == 'nao_declarado'].sample(frac=0.10)
```

```python
amostra_estratificada = pd.concat([branca,parda,preta,amarela,nd,indigena])
amostra_estratificada
```
